```python
import yfinance as yf
import mplfinance as mpf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
symbol = "NVDA"
start = "2023-05-27"
end = "2023-08-27"
data = yf.download(symbol, start=start, end=end)
```

```python
mpf.plot(data, volume=True, type='candle', title=symbol + "Candlestick Chart")
```
